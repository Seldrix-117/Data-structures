```python
import graphviz

class Graph:
    def __init__(self):
        self.graph = {}

    def add_vertex(self, vertex):
        if vertex not in self.graph:
            self.graph[vertex] = []

    def add_edge(self, vertex1, vertex2, weight):
        if vertex1 in self.graph and vertex2 in self.graph:
            # Añadir la arista con el peso
            self.graph[vertex1].append((vertex2, weight))
            self.graph[vertex2].append((vertex1, weight))  # Grafo no dirigido

    def remove_vertex(self, vertex):
        if vertex in self.graph:
            del self.graph[vertex]
            for v in self.graph:
                self.graph[v] = [n for n in self.graph[v] if n[0] != vertex]

    def print_adjacency_list(self):
        for vertex in self.graph:
            edges = ', '.join([f"{neighbor} (peso {weight})" for neighbor, weight in self.graph[vertex]])
            print(f"{vertex}: {edges}")

    def print_adjacency_matrix(self):
        vertices = sorted(self.graph.keys())
        n = len(vertices)
        matrix = [[0] * n for _ in range(n)]
        for i, v1 in enumerate(vertices):
            for neighbor, weight in self.graph[v1]:
                j = vertices.index(neighbor)
                matrix[i][j] = weight  # Guardar el peso en la matriz
        for row in matrix:
            print(row)

    def generate_graphviz_image(self, filename="graph"):
        dot = graphviz.Graph(comment='Undirected Weighted Graph')
        for vertex in self.graph:
            dot.node(str(vertex))
        for vertex in self.graph:
            for neighbor, weight in self.graph[vertex]:
                if vertex < neighbor:  # Evitar añadir aristas dos veces
                    dot.edge(str(vertex), str(neighbor), label=str(weight))
        dot.render(filename, format='png', view=True)

def menu():
    graph = Graph()

    while True:
        print("\n--- Menú de Grafos No Dirigidos Ponderados ---")
        print("1. Añadir vértice")
        print("2. Añadir arista")
        print("3. Eliminar vértice")
        print("4. Mostrar lista de adyacencia")
        print("5. Mostrar matriz de adyacencia")
        print("6. Generar imagen del grafo")
        print("7. Salir")

        choice = input("Selecciona una opción: ")

        if choice == '1':
            vertex = input("Ingresa el número del vértice a añadir: ")
            if vertex.isdigit():
                graph.add_vertex(int(vertex))
                print(f"Vértice {vertex} añadido.")
            else:
                print("Por favor, ingresa un número válido.")
        elif choice == '2':
            vertex1 = input("Ingresa el primer vértice: ")
            vertex2 = input("Ingresa el segundo vértice: ")
            weight = input("Ingresa el peso de la arista: ")
            if vertex1.isdigit() and vertex2.isdigit() and weight.isdigit():
                graph.add_edge(int(vertex1), int(vertex2), int(weight))
                print(f"Arista entre {vertex1} y {vertex2} con peso {weight} añadida.")
            else:
                print("Por favor, ingresa números válidos.")
        elif choice == '3':
            vertex = input("Ingresa el vértice a eliminar: ")
            if vertex.isdigit():
                graph.remove_vertex(int(vertex))
                print(f"Vértice {vertex} eliminado.")
            else:
                print("Por favor, ingresa un número válido.")
        elif choice == '4':
            print("\nLista de adyacencia:")
            graph.print_adjacency_list()
        elif choice == '5':
            print("\nMatriz de adyacencia:")
            graph.print_adjacency_matrix()
        elif choice == '6':
            filename = input("Ingresa el nombre del archivo para guardar la imagen (sin extensión): ")
            graph.generate_graphviz_image(filename)
            print(f"Imagen del grafo generada como {filename}.png.")
        elif choice == '7':
            print("Saliendo...")
            break
        else:
            print("Opción no válida. Por favor, selecciona otra.")

if __name__ == "__main__":
    menu()
```

```python
import graphviz

class WeightedDiGraph:
    def __init__(self):
        self.graph = {}

    def add_vertex(self, vertex):
        if vertex not in self.graph:
            self.graph[vertex] = {}

    def add_edge(self, source, destination, weight):
        if source in self.graph and destination in self.graph:
            self.graph[source][destination] = weight

    def remove_vertex(self, vertex):
        if vertex in self.graph:
            del self.graph[vertex]
            for v in self.graph:
                if vertex in self.graph[v]:
                    del self.graph[v][vertex]

    def print_adjacency_list(self):
        for vertex in self.graph:
            print(f"{vertex}: {self.graph[vertex]}")

    def print_adjacency_matrix(self):
        vertices = sorted(self.graph.keys())
        num_vertices = len(vertices)

        matrix = [[0] * num_vertices for _ in range(num_vertices)]
        vertex_map = {vertex: index for index, vertex in enumerate(vertices)}

        for source in self.graph:
            for destination, weight in self.graph[source].items():
                matrix[vertex_map[source]][vertex_map[destination]] = weight

        print("Matriz de adyacencia:")
        for row in matrix:
            print(row)

    def visualize_graph(self, filename="graph.png"):
        dot = graphviz.Digraph(format='png')

        for vertex in self.graph:
            dot.node(str(vertex))

        for source in self.graph:
            for destination, weight in self.graph[source].items():
                dot.edge(str(source), str(destination), label=str(weight))

        dot.render(filename, view=True)

def menu():
    graph = WeightedDiGraph()

    while True:
        print("\n--- Menú de Grafo Dirigido Ponderado ---")
        print("1. Añadir vértice")
        print("2. Añadir arista")
        print("3. Eliminar vértice")
        print("4. Mostrar lista de adyacencia")
        print("5. Mostrar matriz de adyacencia")
        print("6. Visualizar grafo")
        print("7. Salir")

        choice = input("Selecciona una opción: ")

        if choice == '1':
            vertex = input("Ingresa el nombre del vértice a añadir: ")
            graph.add_vertex(vertex)
            print(f"Vértice '{vertex}' añadido.")
        elif choice == '2':
            source = input("Ingresa el vértice de origen: ")
            destination = input("Ingresa el vértice de destino: ")
            weight = input("Ingresa el peso de la arista: ")
            if weight.isdigit():
                graph.add_edge(source, destination, int(weight))
                print(f"Arista de '{source}' a '{destination}' con peso {weight} añadida.")
            else:
                print("Por favor, ingresa un peso válido.")
        elif choice == '3':
            vertex = input("Ingresa el vértice a eliminar: ")
            graph.remove_vertex(vertex)
            print(f"Vértice '{vertex}' eliminado.")
        elif choice == '4':
            print("\nLista de adyacencia:")
            graph.print_adjacency_list()
        elif choice == '5':
            graph.print_adjacency_matrix()
        elif choice == '6':
            filename = input("Ingresa el nombre del archivo para guardar la imagen (sin extensión): ")
            graph.visualize_graph(filename)
            print(f"Imagen del grafo generada como {filename}.png.")
        elif choice == '7':
            print("Saliendo...")
            break
        else:
            print("Opción no válida. Por favor, selecciona otra.")

if __name__ == "__main__":
    menu()
6
```

```python
import graphviz

class Grafo:
    def __init__(self, dirigido=False, con_peso=False):
        self.dirigido = dirigido
        self.con_peso = con_peso
        self.vertices = set()
        self.aristas = set()  # (origen, destino, peso) o (origen, destino)

    def agregar_vertice(self, vertice):
        self.vertices.add(vertice)

    def agregar_arista(self, origen, destino, peso=1):
        if self.con_peso:
            self.aristas.add((origen, destino, peso))
            if not self.dirigido:
                self.aristas.add((destino, origen, peso))
        else:
            self.aristas.add((origen, destino))
            if not self.dirigido:
                self.aristas.add((destino, origen))

    def eliminar_vertice(self, vertice):
        if vertice in self.vertices:
            self.vertices.remove(vertice)
            if self.con_peso:
                self.aristas = set([(origen, destino, peso) for origen, destino, peso in self.aristas if origen != vertice and destino != vertice])
            else:
                self.aristas = set([(origen, destino) for origen, destino in self.aristas if origen != vertice and destino != vertice])

    def eliminar_arista(self, origen, destino):
        if self.con_peso:
            self.aristas = set([(o, d, p) for o, d, p in self.aristas if not (o == origen and d == destino)])
            if not self.dirigido:
                self.aristas = set([(o, d, p) for o, d, p in self.aristas if not (o == destino and d == origen)])
        else:
            self.aristas = set([(o, d) for o, d in self.aristas if not (o == origen and d == destino)])
            if not self.dirigido:
                self.aristas = set([(o, d) for o, d in self.aristas if not (o == destino and d == origen)])

    def imprimir_matriz_adyacencia(self):
        matriz = [[0] * len(self.vertices) for _ in range(len(self.vertices))]
        vertices_lista = list(self.vertices)
        for arista in self.aristas:
            if self.con_peso:
                origen, destino, peso = arista
            else:
                origen, destino = arista
                peso = 1
            if origen in self.vertices and destino in self.vertices:
                fila = vertices_lista.index(origen)
                columna = vertices_lista.index(destino)
                matriz[fila][columna] = peso
        print("Matriz de Adyacencia:")
        for fila in matriz:
            print(fila)

    def imprimir_lista_adyacencia(self):
        lista_adyacencia = {vertice: [] for vertice in self.vertices}
        for arista in self.aristas:
            if self.con_peso:
                origen, destino, _ = arista
            else:
                origen, destino = arista
            if origen in self.vertices and destino in self.vertices:
                lista_adyacencia[origen].append(destino)
        print("Lista de Adyacencia:")
        for vertice, vecinos in lista_adyacencia.items():
            print(f"{vertice}: {vecinos}")

    def generar_imagen(self, nombre_archivo):
        dot = graphviz.Digraph(format='png') if self.dirigido else graphviz.Graph(format='png')
        for vertice in self.vertices:
            dot.node(str(vertice))
        for arista in self.aristas:
            if self.con_peso:
                origen, destino, peso = arista
                dot.edge(str(origen), str(destino), label=str(peso))
            else:
                origen, destino = arista
                dot.edge(str(origen), str(destino))
        dot.render(nombre_archivo, view=False)

# Menú interactivo
def menu():
    grafo = None

    while True:
        if grafo is None:
            print("\n--- Menú Inicial ---")
            print("1. Crear grafo dirigido con peso")
            print("2. Crear grafo dirigido sin peso")
            print("3. Crear grafo no dirigido con peso")
            print("4. Crear grafo no dirigido sin peso")
            opcion = input("Elige una opción para crear el grafo: ")

            if opcion == "1":
                grafo = Grafo(dirigido=True, con_peso=True)
                print("Grafo dirigido con peso creado.")
            elif opcion == "2":
                grafo = Grafo(dirigido=True, con_peso=False)
                print("Grafo dirigido sin peso creado.")
            elif opcion == "3":
                grafo = Grafo(dirigido=False, con_peso=True)
                print("Grafo no dirigido con peso creado.")
            elif opcion == "4":
                grafo = Grafo(dirigido=False, con_peso=False)
                print("Grafo no dirigido sin peso creado.")
            else:
                print("Opción no válida. Inténtalo de nuevo.")
                continue

        print("\n--- Menú ---")
        print("1. Agregar vértice")
        print("2. Agregar arista")
        print("3. Eliminar vértice")
        print("4. Eliminar arista")
        print("5. Mostrar lista de adyacencia")
        print("6. Mostrar matriz de adyacencia")
        print("7. Generar imagen del grafo")
        print("8. Salir")

        opcion = input("Elige una opción: ")

        if opcion == "1":
            vertice = input("Introduce el vértice: ")
            grafo.agregar_vertice(vertice)
            print(f"Vértice '{vertice}' agregado.")

        elif opcion == "2":
            origen = input("Introduce el vértice de origen: ")
            destino = input("Introduce el vértice de destino: ")
            if grafo.con_peso:
                peso = int(input("Introduce el peso de la arista: "))
                grafo.agregar_arista(origen, destino, peso)
                print(f"Arista de '{origen}' a '{destino}' con peso {peso} agregada.")
            else:
                grafo.agregar_arista(origen, destino)
                print(f"Arista de '{origen}' a '{destino}' agregada.")

        elif opcion == "3":
            vertice = input("Introduce el vértice a eliminar: ")
            grafo.eliminar_vertice(vertice)
            print(f"Vértice '{vertice}' eliminado.")

        elif opcion == "4":
            origen = input("Introduce el vértice de origen: ")
            destino = input("Introduce el vértice de destino: ")
            grafo.eliminar_arista(origen, destino)
            print(f"Arista de '{origen}' a '{destino}' eliminada.")

        elif opcion == "5":
            print("\nLista de Adyacencia:")
            grafo.imprimir_lista_adyacencia()

        elif opcion == "6":
            print("\nMatriz de Adyacencia:")
            grafo.imprimir_matriz_adyacencia()

        elif opcion == "7":
            nombre_archivo = input("Introduce el nombre del archivo para la imagen (sin extensión): ")
            grafo.generar_imagen(nombre_archivo)
            print(f"Grafo guardado como {nombre_archivo}.png")

        elif opcion == "8":
            print("Saliendo del programa...")
            break

        else:
            print("Opción no válida. Inténtalo de nuevo.")

# Iniciar el menú
menu()
```
